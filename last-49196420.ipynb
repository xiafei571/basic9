## Penalty shootout game

```python
import time
from random import *

# User
class User:
    def __init__(self, uid, uname, ST, GK): #Shooter, Goalkeeper
        self.__uid = uid
        self.__uname = uname
        self.ST = ST
        self.GK = GK
    
    def info(self):
        return '-'.join([str(self.__uid), self.__uname])

# Player
class Player:
    def __init__(self, pid, pname, overall):
        self.__pid = pid
        self.__pname = pname
        self.__overall = overall
        
    def info(self):
        return ','.join([str(self.__pid), self.__panme, str(self.__overall)])
    
    def get_overall(self):
        return self.__overall
    

# Game
class Game():
    def __init__(self,u1, u2, rounds):
        self.__game_id = int(time.time())
        self.u1 = u1 # user1
        self.u2 = u2 # user2
        self.__s1 = 0 # score1
        self.__s2 = 0 # score2
        self.__round = 0
        self.__max_round = rounds
        
    def game_over(self): 
#         print(self.__s1, self.__s2)
#         print(self.__round, self.__max_round)
        if self.__round >= self.__max_round:
            if self.__s1 == self.__s2:
                return False
            else:
                return True
        else:
            if self.__s1 == self.__s2:
                return False
            else:
                remaining = self.__max_round - self.__round
                if (self.__s1 - self.__s2) > remaining or (self.__s2 - self.__s1) > remaining:
                    return True
                else:
                    return False
                
    
    def round_go(self):
        self.__round += 1
        GK_ACT_1 = input('Please user1 input the action of GK:')
        ST_ACT_1 = input('Please user2 input the action of ST:')
        print('user2 get:', self.round_one(GK_ACT_1, ST_ACT_1))
        ST_ACT_2 = input('Please user1 input the action of ST:')
        GK_ACT_2 = input('Please user2 input the action of GK:')
        print('user1 get:', self.round_two(GK_ACT_2, ST_ACT_2))
        print('round:',self.__round,', score:',self.__s1,":",self.__s2)
            
        
    def round_one(self, GK_ACT, ST_ACT):
        #u1 -> GK, u2 -> ST
        g = u1.GK.get_overall() / 100
        s = u2.ST.get_overall() / 100
        sr = random() # random 0-1
        if sr > s:
            print('Shooting Missed')
            return 0
        else:
            # Shooting direction
            if GK_ACT != ST_ACT:
                print('Goal!!')
                self.__s2 += 1
                return 1
            else:
                gr = random()
                if gr > g:
                    print('Goal!!')
                    self.__s2 += 1
                    return 1
                else:
                    print('Amazing save!')
                    return 0
        
        
    def round_two(self, GK_ACT, ST_ACT):
        #u1 -> ST, u2 -> GK
        s = u1.ST.get_overall() / 100
        g = u2.GK.get_overall() / 100
        sr = random() # random 0-1
        if sr > s:
            print('Shooting Missed')
            return 0
        else:
            # Shooting direction
            if GK_ACT != ST_ACT:
                print('Goal!!')
                self.__s1 += 1
                return 1
            else:
                gr = random()
                if gr > g:
                    print('Goal!!')
                    self.__s1 += 1
                    return 1
                else:
                    print('Amazing save!')
                    return 0
        
    def show_view(self):
        print('-------------')
        print('| Q | W | E |')
        print('-------------')
        print('| A | S | D |')
        print('-------------')
        
    def show_result(self):
        print(self.u1.info())
        print('score:', self.__s1)
        print(self.u2.info())
        print('score:', self.__s2)
        
def main():
    p1_st = Player(158023, 'L. Messi', 94)
    p1_gk = Player(193080, 'De Gea', 91)

    p2_st = Player(20801, 'CristianoRonaldo', 94)
    p2_gk = Player(200389, 'J. Oblak', 90)

    u1 = User(1001, 'Tom', p1_st, p1_gk)
    u2 = User(1002, 'Jerry', p2_st, p2_gk)

    g1 = Game(u1, u2, 5)
    while not g1.game_over():
        g1.show_view()
        g1.round_go()
        
    g1.show_result()
    
if __name__ == '__main__':
    main()


```

```python
import socket

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect(("127.0.0.1", 9999))

while True:
    # receive msg
    data = s.recv(1024).decode("utf-8")
    print("server data:", data)

    # send msg
    inp = input('client data: ')
    s.send(inp.encode("utf-8"))
    if inp == 'exit()':
        break
s.close()
```
